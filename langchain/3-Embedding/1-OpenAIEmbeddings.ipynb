```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.getenv("OPENAPI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
model = "text-embedding-3-large"
embeddings_model = OpenAIEmbeddings(api_key=api_key, model=model)
```

```python
query = "This is a tutorial on OpenAI embedding"

try:
  embeddings = embeddings_model.embed_query(query)
  print(embeddings)
except Exception as ex:
  print(f"Error - {ex}")
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
```

```python
# Data Ingestion
loader = TextLoader('docs/speech.txt')
docs = loader.load()

# Data Transformation | TextSplitter 
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
)

chunked_docs = text_splitter.split_documents(docs)
```

```python
# Initializing OpenAIEmbeddings via its model (reduced dimensions)
model = "text-embedding-3-large"
embeddings_model_1024 = OpenAIEmbeddings(api_key=api_key, model=model, dimensions=1024)
```

```python
# Using VectorDB - Chroma | Converting and storing embeddings for the docs in single step in vectordb
db = Chroma.from_documents(chunked_docs, embeddings_model_1024)

search_query = "It will be all the easier for us to conduct ourselves as belligerents"

# searching a part of text in vector db
result = db.similarity_search(query)
print(result)
```
