# Ollama Embedding

Using OpenAI Embedding Model - `gemma2:2b`

```python
from langchain_community.embeddings import OllamaEmbeddings
model = "gemma2:2b"
embeddings_model = OllamaEmbeddings(model=model)
```

```python
r1 = embeddings_model.embed_documents([
    "Alpha is the first letter of the greek alphabet",
    "Beta is the second letter of the greek alphabet",
])

r1[0]
```

```python
query = "This is a tutorial on Ollama embedding"

try:
  embeddings = embeddings_model.embed_query(query)
  print(embeddings)
except Exception as ex:
  print(f"Error - {ex}")
```
